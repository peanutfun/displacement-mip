# Flood Hazard Calculation

```python
from pathlib import Path

import numpy as np

from climada_petals.hazard.rf_glofas import RiverFloodInundation, save_file
```

```python
rf = RiverFloodInundation()

output_path = Path("./data/generated/sudan/hazard")
years = np.arange(2008, 2022 + 1)
for year in years.flat:
    rf.download_reanalysis("Sudan", year, preprocess=lambda x: x.max(dim="time"))
    ds_flood = rf.compute()
    save_file(ds_flood, output_path / f"sudan-{year}.nc")
    rf.clear_cache()
```

```python
from pathlib import Path

import numpy as np
import xarray as xr

from climada_petals.hazard.rf_glofas import RiverFloodInundation, save_file

countries = ["Sudan", "Ethiopia", "Somalia"]

with xr.open_dataarray("data/discharge-2023.grib", chunks="auto") as arr:
    for country in countries:
        # Subselection of arr for country
        # numpy ndarray of [xmin, ymin, xmax, ymax]
        bounds = get_country_geometries(country_to_iso(country)).total_bounds
        discharge = arr.sel(
            latitude=slice(bounds[3], bounds[1]), longitude=slice(bounds[0], bounds[2])
        ).max(dim="time").expand_dims("time")

        # Compute rf
        rf = RiverFloodInundation()
        ds_flood = rf.compute(discharge=discharge)

        # save rf
        country = country.lower()
        output_path = Path(f"./data/generated/{country}/hazard")
        save_file(ds_flood, output_path / f"{country}-2023.nc")
        rf.clear_cache()
```
