# Flood Hazard Calculation

```python
from pathlib import Path

import numpy as np

from climada_petals.hazard.rf_glofas import RiverFloodInundation, save_file
```

```python
rf = RiverFloodInundation()

output_path = Path("./data/generated/sudan/hazard")
years = np.arange(2008, 2022 + 1)
for year in years.flat:
    rf.download_reanalysis("Sudan", year, preprocess=lambda x: x.max(dim="time"))
    ds_flood = rf.compute()
    save_file(ds_flood, output_path / f"sudan-{year}.nc")
    rf.clear_cache()
```
