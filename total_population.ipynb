# Compute Total Population from Exposure

```python
import pandas as pd
from impact_calc import load_worldpop, YEAR_RANGE_DEFAULT
from climada.util.coordinates import country_to_iso


def total_population(exposure):
    return exposure.gdf["value"].sum()


countries = ("ethiopia", "somalia", "sudan")

df_pop = pd.concat(
    [
        pd.Series(
            [
                total_population(load_worldpop(country, year))
                for year in YEAR_RANGE_DEFAULT
            ],
            index=YEAR_RANGE_DEFAULT,
            name=country_to_iso(country),
        )
        for country in countries
    ],
    axis=1,
)
df_pop = df_pop.astype("int")
df_pop.index.rename("Year", inplace=True)
df_pop.columns.rename("Country", inplace=True)
df_pop
```

```python
df_pop.to_csv("total_population.csv")
```
